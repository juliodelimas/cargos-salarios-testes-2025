### **1. Introdução e Objetivos**

Este estudo realiza uma análise longitudinal detalhada sobre a evolução do perfil técnico, da remuneração e das desigualdades estruturais no mercado brasileiro de Testes e Qualidade de Software (QA) entre os anos de 2023 e 2025. #### Objetivos principais:

1. **Evolução de Perfis Técnicos e Salariais:**

### **4.3 Clusterização**

A clusterização foi conduzida por meio da combinação de **redução de dimensionalidade via PCA** (2 componentes principais) e agrupamento com **KMeans (k=4)**, utilizando como variáveis explicativas:

- `salario` (valor bruto mensal);
- `senioridade` (categorias padronizadas: júnior, pleno, sênior, etc.);

- `modelo_trabalho` (presencial, híbrido, remoto).

Após padronização e codificação de variáveis categóricas, os dados foram projetados em duas dimensões principais que capturam a maior variância possível dos perfis profissionais. O algoritmo KMeans identificou **quatro grupos** distintos de profissionais com padrões salariais e estruturais específicos:

---

#### **Cluster 0 — Perfil Heterogêneo de Amplitude Salarial**
- É o maior cluster em termos de densidade e dispersão;
- Composto por profissionais com **ampla variação salarial** (de júnior a sênior);
- Representa uma base populacional distribuída por diversas regiões e modelos de trabalho;
- Pode incluir perfis em transição de carreira ou com múltiplas combinações de atributos.

---

#### **Cluster 1 — Profissionais Intermediários e Remotos**
- Concentra-se em profissionais com **modelo de trabalho remoto ou híbrido**, especialmente em regiões mais desenvolvidas (Sudeste, Sul);
- Faixa salarial superior à média, com predominância de **senioridade plena ou sênior**;
- Indica um perfil que se beneficia da flexibilidade geográfica e da valorização técnica no mercado atual.

---

#### **Cluster 2 — Perfis de Entrada Presenciais**
- Predominantemente formado por profissionais de **nível júnior**, residentes em **regiões com menor remuneração média** (Norte, Nordeste);
- Associado a **modelos presenciais de atuação** e faixas salariais mais baixas;
- Representa um grupo que ainda não internalizou os ganhos de mobilidade ou capacitação técnica avançada.

---

#### **Cluster 3 — Grupo de Transição ou Outlier Estrutural**
- Menor em volume e mais concentrado nos extremos dos componentes principais;
- Possivelmente representa perfis com **comportamento atípico**, como salários acima da média em regiões de baixa remuneração ou níveis sênior com modelo presencial;
- Pode indicar uma população de **transição entre clusters dominantes** ou **outliers com padrão misto**.

---

> **Nota**: Esta clusterização não considera ferramentas (ex: Cypress, Playwright, uso de IA) nem o regime de contratação (CLT, PJ), uma vez que essas variáveis não foram incluídas no modelo original.

#### **4.4 Recuperação Pós-Layoffs**

Os dados dos anos de 2023, 2024 e 2025 foram integrados, totalizando 4175 registros e 39 colunas. Entretanto, como observado, há duplicidade de campos por conta de diferentes formulários utilizados em cada ano (ex: múltiplas colunas para salário, senioridade, modelo de trabalho etc.). Desse modo, passará por uma limpeza nos dados.

````python
# Pipeline de limpeza
class DataPipeline:
    def __init__(self, df: pd.DataFrame):
        self.df = df.copy()

    def _merge_salario(self):
        self.df["salario"] = self.df["Qual é o seu salário mensal bruto em Reais?"].combine_first(
            self.df["Qual é o seu salário mensal bruto? (Informe a moeda)"]
        ).combine_first(
            self.df["Qual é o seu salário bruto?"]
        )
        return self

    def _merge_senioridade(self):
        self.df["senioridade"] = self.df["Qual é sua senioridade?"].combine_first(
            self.df["Qual é a sua senioridade?"]
        )
        return self

    def _merge_modelo_trabalho(self):
        self.df["modelo_trabalho"] = self.df["Modelo de Atuação"].combine_first(
            self.df["Trabalho remoto"]
        )
        return self

    def _merge_genero(self):
        self.df["genero"] = self.df["Qual é seu gênero?"]
        return self

    def _merge_regiao(self):
        self.df["regiao"] = self.df["Em qual região do Brasil você mora?"].combine_first(
            self.df["Em qual região do Brasil você vive atualmente?"]
        )
        return self

    def _normalize_strings(self):
        str_cols = ["senioridade", "modelo_trabalho", "genero", "regiao"]
        for col in str_cols:
            self.df[col] = self.df[col].astype(str).str.strip().str.lower()
        return self

    def _filter_salario_validos(self, min_val=500, max_val=80000):
        self.df = self.df[(self.df["salario"] >= min_val) & (self.df["salario"] <= max_val)]
        return self

    def run(self):
        return (
            self._merge_salario()
                ._merge_senioridade()
                ._merge_modelo_trabalho()
                ._merge_genero()
                ._merge_regiao()
                ._normalize_strings()
                ._filter_salario_validos()
                .df
        )

# Aplicando o pipeline aos dados
dados_limpos = DataPipeline(dados_completos).run()
````

A limpeza consolidou as principais colunas sob nomes padronizados:

- `salario`: valor bruto padronizado;
- `senioridade`: unificada e normalizada;
- `modelo_trabalho`: inclui remoto, híbrido, presencial;
- `genero` e `regiao`: prontos para análises demográficas.

````python
# Filtrar colunas relevantes para análise
dados_salario = dados_limpos[["Ano", "salario", "regiao", "senioridade"]].dropna()

# Agrupar para estatísticas básicas
salario_por_ano_regiao = (
    dados_salario
    .groupby(["Ano", "regiao"])
    .agg(media_salario=("salario", "mean"),
         mediana_salario=("salario", "median"),
         std_salario=("salario", "std"),
         amostra=("salario", "count"))
    .reset_index()
)

# Boxplot por Região e Ano
plt.figure(figsize=(14, 6))
sns.boxplot(data=dados_salario, x="regiao", y="salario", hue="Ano", palette=pastel_colors)
plt.title("Distribuição Salarial por Região e Ano")
plt.xlabel("Região do Brasil")
plt.ylabel("Salário Bruto Mensal (BRL)")
plt.legend(title="Ano")
plt.tight_layout()
plt.show()
````

### Gráfico Box-Plot: Distribuição Salarial por Região e Ano

O gráfico mostra:

- **Norte e Nordeste** apresentam distribuições mais concentradas e com menores salários;

````python
# Teste ANOVA por região para detectar diferenças significativas de salário entre os anos
anova_results = {}

# Iterar por cada região e aplicar ANOVA se houver dados em pelo menos dois anos com amostras válidas
for regiao in dados_salario["regiao"].dropna().unique():
    dados_regiao = dados_salario[dados_salario["regiao"] == regiao]
    grupos = [
        grupo["salario"].values
        for _, grupo in dados_regiao.groupby("Ano")
        if len(grupo) > 1
    ]

    # Só aplicamos ANOVA se houver pelo menos dois grupos válidos
    if len(grupos) >= 2:
        f_valor, p_valor = stats.f_oneway(*grupos)
        anova_results[regiao] = {"F": round(f_valor, 2), "p-valor": round(p_valor, 4)}

# Exibir resultados como DataFrame para leitura
anova_df = pd.DataFrame.from_dict(anova_results, orient="index").sort_values(by="p-valor")

anova_df
````

### **Análise de Variância (ANOVA) por Região – Salário (2023–2025)**

A análise ANOVA (Análise de Variância Unifatorial) foi aplicada para verificar se as **diferenças nas médias salariais ao longo dos anos (2023, 2024 e 2025)** são estatisticamente significativas dentro de cada uma das cinco regiões brasileiras. O teste F compara a variância intra-grupo (dentro de cada ano) com a variância inter-grupo (entre os anos), assumindo a hipótese nula de que **as médias salariais ao longo dos anos são iguais** dentro de cada região.

Abaixo segue a interpretação dos resultados por região:

| Região        | Estatística F | Valor-p (p-value) | Interpretação da Tabela |
|---------------|---------------|-------------------|------------------------|
| **Sudeste**       | **5.20**         | **0.0056**             | Diferença altamente significativa. Há evidência estatística de que as médias salariais variaram entre os anos, com alta confiabilidade (nível de significância α = 0.01). |
| **Nordeste**      | **3.15**         | **0.0433**             | Diferença estatisticamente significativa em nível convencional (α = 0.05). Indica possível variação salarial ao longo dos anos, embora com menor robustez que o Sudeste. |
| **Centro-Oeste**  | 1.49             | 0.2270                | Diferença **não significativa**. As médias salariais ao longo dos anos podem ser consideradas estatisticamente equivalentes dentro da região. |
| **Sul**           | 0.92             | 0.3975                | Sem evidência de variação estatisticamente relevante entre os anos. A média salarial permaneceu estável. |
| **Norte**         | 0.29             | 0.7454                | Variações salariais nos anos analisados não foram significativas. A hipótese nula não é rejeitada com alto grau de confiança. |

---

### **Conclusões dos testes ANOVA**

- A **região Sudeste** apresentou a maior estatística F (5.20) com **p < 0.01**, o que significa que há **fortes evidências estatísticas de variação salarial real entre os anos**. Essa variação pode estar associada ao crescimento de regimes PJ, adoção de ferramentas modernas (ex: Cypress, IA) ou maior penetração de trabalho remoto com contratos internacionais.
  
- A **região Nordeste** também apresenta variação estatisticamente significativa (p = 0.0433), embora com menor magnitude. Pode refletir o impacto de políticas locais de capacitação ou reestruturações contratuais no período pós-layoffs.
  
- Nas demais regiões (Centro-Oeste, Sul e Norte), os valores-p estão bem acima de 0.05, indicando que **não há evidências robustas para rejeitar a hipótese de estabilidade salarial** ao longo do período. Isso sugere que o comportamento salarial foi mais estático ou menos impactado por transformações estruturais nesses locais.

### **No que isso implica?**

- O resultado da ANOVA serve como **base quantitativa para priorizar políticas de ajuste salarial e capacitação técnica por região**.
- O fato de apenas duas regiões apresentarem significância estatística reforça a **concentração geográfica das transformações salariais e técnicas**, o que pode ser tratado como um **indicador de desigualdade regional em evolução tecnológica**.

````python
# Selecionar as colunas de ferramentas técnicas com ano
colunas_competencia = [
    "Ano",
    "(Opcional) Quais ferramentas de automação de testes você utiliza na empresa onde você trabalha atualmente?",
    "(Opcional) Quais linguagens de programação são utilizadas para automação de testes na empresa onde você trabalha?"
]

# Filtrar dataset
dados_competencias = dados_limpos[colunas_competencia].dropna(how='all', subset=colunas_competencia[1:])

# Função para processar texto e remover stopwords
def extrair_palavras(textos: List[str]) -> List[str]:
    palavras = []
    stopwords_pt = set(stopwords.words("portuguese"))
    for texto in textos:
        if isinstance(texto, str):
            tokens = re.split(r"[,\s;/]+", texto.lower())
            palavras.extend([t.strip() for t in tokens if t.strip() not in stopwords_pt and len(t.strip()) > 2])
    return palavras

# Gerar nuvens de palavras por ano
wordclouds = {}
for ano in sorted(dados_competencias["Ano"].unique()):
    dados_ano = dados_competencias[dados_competencias["Ano"] == ano]
    palavras = extrair_palavras(
        dados_ano[colunas_competencia[1]].dropna().tolist() +
        dados_ano[colunas_competencia[2]].dropna().tolist()
    )
    frequencias = Counter(palavras)
    wc = WordCloud(width=1000, height=500, background_color="white", colormap="Pastel1").generate_from_frequencies(frequencias)
    wordclouds[ano] = wc

# Exibir nuvens lado a lado (ou única)
fig, axes = plt.subplots(1, len(wordclouds), figsize=(6 * len(wordclouds), 6))

# Ajuste: transforma em lista se só houver um gráfico
if len(wordclouds) == 1:
    axes = [axes]

# Loop de exibição
for i, (ano, wc) in enumerate(sorted(wordclouds.items())):
    axes[i].imshow(wc, interpolation="bilinear")
    axes[i].axis("off")
    axes[i].set_title(f"Competências Técnicas - {ano}", fontsize=14)

plt.tight_layout()
plt.show()
````

````python
# Seleção e pré-processamento de variáveis
variaveis_cluster = dados_limpos[["salario", "senioridade", "regiao", "modelo_trabalho"]].dropna()
variaveis_dummies = pd.get_dummies(variaveis_cluster, columns=["senioridade", "regiao", "modelo_trabalho"], drop_first=True)

# Padronização
scaler = StandardScaler()
X_scaled = scaler.fit_transform(variaveis_dummies)

# Redução de dimensionalidade com PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# KMeans com 4 clusters
kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
labels = kmeans.fit_predict(X_scaled)

# DataFrame com componentes principais e rótulo do cluster
df_clusters = pd.DataFrame(X_pca, columns=["PC1", "PC2"])
df_clusters["Cluster"] = labels

# Plotagem dos clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_clusters, x="PC1", y="PC2", hue="Cluster", palette="pastel", s=60)
plt.title("Clusterização de Perfis Profissionais (PCA + KMeans)")
plt.xlabel("Componente Principal 1")
plt.ylabel("Componente Principal 2")
plt.legend(title="Cluster")
plt.grid(True)
plt.tight_layout()
plt.show()
````

- **Algoritmo KMeans (k=4)** aplicado sobre a matriz original para formação dos clusters, utilizando critério de inércia mínima (`k=4` foi definido empiricamente por meio de inspeção visual dos centróides e variabilidade dos grupos).

- O grupo masculino apresenta **concentração maior de salários elevados**, o que pode estar relacionado à **retomada de contratos PJ ou reemprego em posições técnicas avançadas**.
- Mulheres, mesmo com dispersão visível, **não acessam a faixa de outliers superiores** com a mesma frequência, o que pode refletir **restrições contratuais, geográficas ou culturais** na recontratação.

#### **3.3 Interações Esperadas com Senioridade e Regime**

- Espera-se que ao cruzar com a variável `senioridade`, o impacto sobre mulheres seja ainda mais expressivo nas faixas “sênior” e “coordenador(a)”.
- A ausência de regime de contratação no gráfico oculta um fator-chave: **profissionais PJ tendem a se recuperar mais rapidamente**, o que pode explicar parte da diferença entre os grupos.

---

### **4. Implicações**

- Os resultados indicam que **eventos sistêmicos como layoffs tendem a amplificar desigualdades salariais já existentes** entre gêneros.
- **Processos de recontratação e reinserção** estão sujeitos a vieses estruturais que penalizam de forma desproporcional profissionais mulheres, especialmente nos níveis técnicos intermediários e avançados.
- A estagnação salarial entre mulheres afetadas, combinada à dispersão masculina, sugere que **estratégias de recuperação e retenção pós-crise foram assimétricas**.

````python
# Extração de ferramentas e conhecimento técnico
col_ferramentas = "(Opcional) Quais ferramentas de automação de testes você utiliza na empresa onde você trabalha atualmente?"
col_ia_testes = "(Opcional) Qual relação você possui com Inteligência Artificial e Testes na empresa onde você trabalha arualmente?"

# Criar coluna binária: conhece ferramentas modernas (Cypress ou Playwright)
def conhece_ferramenta_moderna(texto):
    if isinstance(texto, str):
        texto = texto.lower()
        return any(x in texto for x in ["cypress", "playwright"])
    return False

dados_completos["conhece_ferramenta_moderna"] = dados_completos[col_ferramentas].apply(conhece_ferramenta_moderna)

# Criar coluna binária: tem alguma interação com IA
def tem_ia(texto):
    if isinstance(texto, str):
        texto = texto.lower()
        return "sim" in texto or "automatização" in texto or "machine learning" in texto
    return False

dados_completos["usa_ia_testes"] = dados_completos[col_ia_testes].apply(tem_ia)

# Consolidar layoff
col_layoff_impacto = [
    "Você foi afetado pela onde de demissões em massa nos últimos 6 meses?",
    "Você foi afetado pela onda de demissões em massa nos últimos 6 meses?",
    "(Opcional) Você foi atingido por algum lay off nos últimos 12 meses?"
]

dados_completos["afetado_layoff"] = (
    dados_completos[col_layoff_impacto[0]]
    .combine_first(dados_completos[col_layoff_impacto[1]])
    .combine_first(dados_completos[col_layoff_impacto[2]])
).astype(str).str.strip().str.lower()

# Filtrar apenas colunas relevantes
df_tecnico_layoff = dados_completos[[
    "Ano", "salario", "afetado_layoff", "conhece_ferramenta_moderna", "usa_ia_testes"
]].dropna()

# Classificação binária: afetado ou não
df_tecnico_layoff["afetado"] = df_tecnico_layoff["afetado_layoff"].apply(lambda x: "sim" if "sim" in x else "não")

# Agrupar por status de layoff
resumo_tecnico = df_tecnico_layoff.groupby("afetado").agg(
    total=("Ano", "count"),
    pct_ferramenta_moderna=("conhece_ferramenta_moderna", "mean"),
    pct_usa_ia=("usa_ia_testes", "mean")
).reset_index()

# Converter para porcentagens
resumo_tecnico["pct_ferramenta_moderna"] = (resumo_tecnico["pct_ferramenta_moderna"] * 100).round(1)
resumo_tecnico["pct_usa_ia"] = (resumo_tecnico["pct_usa_ia"] * 100).round(1)

display(resumo_tecnico)
````

````python
# Consolidando coluna de senioridade
dados_completos["senioridade"] = dados_completos["Qual é sua senioridade?"].combine_first(
    dados_completos["Qual é a sua senioridade?"]
).astype(str).str.strip().str.lower()

# Reclassificando afetado
dados_completos["afetado_layoff"] = (
    dados_completos["Você foi afetado pela onde de demissões em massa nos últimos 6 meses?"]
    .combine_first(dados_completos["Você foi afetado pela onda de demissões em massa nos últimos 6 meses?"])
    .combine_first(dados_completos["(Opcional) Você foi atingido por algum lay off nos últimos 12 meses?"])
).astype(str).str.strip().str.lower()

# Classificar se foi afetado ou não
dados_completos["afetado"] = dados_completos["afetado_layoff"].apply(lambda x: "sim" if "sim" in x else "não")

# Análise da proporção de profissionais afetados por senioridade
df_senioridade = dados_completos[["senioridade", "afetado"]].dropna()
resumo_senioridade = (
    df_senioridade
    .groupby(["senioridade", "afetado"])
    .size()
    .unstack(fill_value=0)
)

# Calcular porcentagem de afetados dentro de cada senioridade
resumo_senioridade["% afetado"] = (resumo_senioridade["sim"] / (resumo_senioridade["sim"] + resumo_senioridade["não"]) * 100).round(1)
resumo_senioridade = resumo_senioridade.sort_values(by="% afetado", ascending=False).reset_index()

display(resumo_senioridade)
````

````python
# Preparar análise complementar: salário médio entre júnior e sênior com base em uso de ferramentas técnicas
df_salario_tecnico = dados_completos[[
    "salario", "senioridade", col_ferramentas, col_ia_testes
]].dropna()

# Filtrar para apenas júnior e sênior
df_salario_tecnico["senioridade"] = df_salario_tecnico["senioridade"].str.lower()
df_salario_tecnico = df_salario_tecnico[df_salario_tecnico["senioridade"].isin(["júnior", "pleno", "sênior"])]

# Colunas binárias de conhecimento técnico
df_salario_tecnico["conhece_cypress"] = df_salario_tecnico[col_ferramentas].astype(str).str.lower().str.contains("cypress")
df_salario_tecnico["conhece_playwright"] = df_salario_tecnico[col_ferramentas].astype(str).str.lower().str.contains("playwright")
df_salario_tecnico["usa_ia"] = df_salario_tecnico[col_ia_testes].astype(str).str.lower().str.contains("sim|machine learning|inteligência|automatiza")

# Agrupamento
resumo_salario = (
    df_salario_tecnico
    .groupby(["senioridade", "conhece_cypress", "conhece_playwright", "usa_ia"])
    .agg(media_salario=("salario", "mean"), qtd=("salario", "count"))
    .reset_index()
    .sort_values(by="media_salario", ascending=False)
)

resumo_salario["media_salario"] = resumo_salario["media_salario"].round(2)

display(resumo_salario)
````

````python
# Regressão salarial para estimar impacto das variáveis controlando para gênero, senioridade, ferramentas e modelo de trabalho
# Seleção e limpeza de variáveis
df_regressao = dados_completos[[
    "salario",
    "senioridade",
    "genero",
    "modelo_trabalho",
    col_ferramentas,
    col_linguagens,
    col_ia_testes
]].dropna()

# Filtrar salários válidos
df_regressao = df_regressao[(df_regressao["salario"] >= 500) & (df_regressao["salario"] <= 80000)]

# Normalizar variáveis categóricas
df_regressao["senioridade"] = df_regressao["senioridade"].str.strip().str.lower()
df_regressao["genero"] = df_regressao["genero"].str.strip().str.lower()
df_regressao["modelo_trabalho"] = df_regressao["modelo_trabalho"].str.strip().str.lower()

# Variáveis binárias para ferramentas
df_regressao["cypress"] = df_regressao[col_ferramentas].astype(str).str.lower().str.contains("cypress")
df_regressao["playwright"] = df_regressao[col_ferramentas].astype(str).str.lower().str.contains("playwright")
df_regressao["selenium"] = df_regressao[col_ferramentas].astype(str).str.lower().str.contains("selenium")
df_regressao["usa_ia"] = df_regressao[col_ia_testes].astype(str).str.lower().str.contains("sim|inteligência|machine learning|automatiza")

# Fórmula de regressão com variáveis categóricas + técnicas
formula = "salario ~ C(senioridade) + C(genero) + C(modelo_trabalho) + cypress + playwright + selenium + usa_ia"

# Design matrix
y, X = patsy.dmatrices(formula, data=df_regressao, return_type="dataframe")

# Modelo OLS
modelo = sm.OLS(y, X).fit()

# Resumo do modelo
resumo_modelo = modelo.summary2().tables[1]  # Tabela com coeficientes
resumo_modelo_impacto = resumo_modelo[["Coef.", "P>|t|"]].round(4)

display(resumo_modelo_impacto)
````

### **1. Fundamentos da Regressão Aplicada**

O modelo possui a seguinte estrutura estatística:

```plaintext
salario ~ C(senioridade) + C(genero) + C(modelo_trabalho) + cypress + playwright + selenium + usa_ia
```

Ou seja, estima-se o salário **controlando os efeitos cruzados** dessas variáveis, permitindo observar **o impacto marginal de cada fator** enquanto os demais são mantidos constantes.

A variável de referência (base) para cada fator categórico é automaticamente omitida, como no caso de `senioridade`, onde a base possivelmente é "estagiário" ou "estágio".

````python
# Preparar dados para regressão quantílica
df_quantil = dados_completos[[
    "salario", "senioridade", "genero", "modelo_trabalho", "regime_contratacao",
    col_ferramentas, col_ia_testes
]].dropna()

# Filtrar outliers para evitar distorção
df_quantil = df_quantil[df_quantil["salario"] <= 40000]

# Normalizações
df_quantil["senioridade"] = df_quantil["senioridade"].str.strip().str.lower()
df_quantil["genero"] = df_quantil["genero"].str.strip().str.lower()
df_quantil["modelo_trabalho"] = df_quantil["modelo_trabalho"].str.strip().str.lower()
df_quantil["regime_contratacao"] = df_quantil["regime_contratacao"].str.strip().str.lower()

# Binários técnicos
df_quantil["cypress"] = df_quantil[col_ferramentas].astype(str).str.lower().str.contains("cypress")
df_quantil["playwright"] = df_quantil[col_ferramentas].astype(str).str.lower().str.contains("playwright")
df_quantil["usa_ia"] = df_quantil[col_ia_testes].astype(str).str.lower().str.contains("sim|automatiza|inteligência|machine learning")

# Criar fórmula para regressão quantílica
formula = "salario ~ C(senioridade) + C(genero) + C(modelo_trabalho) + C(regime_contratacao) + cypress + playwright + usa_ia"

# Rodar modelo para quantis 0.25, 0.5, 0.75
quantis = [0.25, 0.5, 0.75]
resultados_quantis = {}

for q in quantis:
    modelo_q = smf.quantreg(formula, data=df_quantil).fit(q=q)
    coef = modelo_q.params.round(2)
    resultados_quantis[f"Q{int(q*100)}"] = coef

# Consolidar em DataFrame
df_quantil_coef = pd.DataFrame(resultados_quantis).fillna(0)

display(df_quantil_coef)
````

````python
# Preparar dados para modelagem temporal baseada em atributos agregados
df_forecast = dados_completos[[
    "Ano", "salario", "senioridade", "modelo_trabalho", "regime_contratacao",
    "genero", col_ferramentas, col_ia_testes
]].dropna()

df_forecast = df_forecast[df_forecast["salario"] <= 40000]

# Normalização
df_forecast["senioridade"] = df_forecast["senioridade"].str.strip().str.lower()
df_forecast["modelo_trabalho"] = df_forecast["modelo_trabalho"].str.strip().str.lower()
df_forecast["regime_contratacao"] = df_forecast["regime_contratacao"].str.strip().str.lower()
df_forecast["genero"] = df_forecast["genero"].str.strip().str.lower()

# Features binárias
df_forecast["usa_cypress"] = df_forecast[col_ferramentas].astype(str).str.lower().str.contains("cypress")
df_forecast["usa_playwright"] = df_forecast[col_ferramentas].astype(str).str.lower().str.contains("playwright")
df_forecast["usa_ia"] = df_forecast[col_ia_testes].astype(str).str.lower().str.contains("inteligência|automatiza|machine learning|sim")

# Agrupamento por ano e variáveis categóricas para média salarial
agrupado = (
    df_forecast
    .groupby(["Ano", "senioridade", "modelo_trabalho", "regime_contratacao", "genero", "usa_cypress", "usa_playwright", "usa_ia"])
    .agg(media_salario=("salario", "mean"))
    .reset_index()
)

# Separar dados de 2023 a 2025 para treino e prever 2026
X = agrupado[agrupado["Ano"] < 2026].copy()
y = X.pop("media_salario")

# Salvar últimos grupos únicos para prever o mesmo padrão em 2026
grupos_2025 = X[X["Ano"] == 2025].drop(columns=["Ano"]).copy()
grupos_2026 = grupos_2025.copy()
grupos_2026["Ano"] = 2026

# Pré-processamento com OneHotEncoder
cat_cols = ["senioridade", "modelo_trabalho", "regime_contratacao", "genero"]
bin_cols = ["usa_cypress", "usa_playwright", "usa_ia"]
num_cols = ["Ano"]

preprocessor = ColumnTransformer(transformers=[
    ("cat", OneHotEncoder(handle_unknown="ignore"), cat_cols),
    ("passthrough", "passthrough", num_cols + bin_cols)
])

# Pipeline com Random Forest
model = Pipeline(steps=[
    ("preprocess", preprocessor),
    ("regressor", RandomForestRegressor(n_estimators=100, random_state=42))
])

# Treinar o modelo
model.fit(X, y)

# Prever para 2026
y_pred_2026 = model.predict(grupos_2026)

# Consolidar resultados
grupos_2026["salario_previsto_2026"] = np.round(y_pred_2026, 2)

display(grupos_2026)
````
